# 3 - Hopfield Network

```python
import numpy as np
import matplotlib.pyplot as plt
import itertools

class HopfieldNetwork:
    def __init__(self):
        self.weights = 0
        self.pattern = 0
        self.targets = 0

    def sign(self, data):
        return np.where(data >= 0, 1, -1)

    def transfer(self, data):
        output = np.dot(self.weights, data)
        return self.sign(output)

    def outer_product(self, one, two):
        self.weights = np.outer(one, two)
        return np.outer(one, two)

    def learn_weights(self, targets):
        #self.pattern = data
        self.targets = targets
        #self.weights = np.dot(self.targets, np.transpose(self.targets))
        self.weights = np.dot(targets.transpose(), targets)
        #for i in range(self.weights.shape[0]):
        #    self.weights[i,i] = 0
            
    def learn_weights_sparse(self, patterns, activity):
        self.weights = np.zeros((patterns.shape[1],patterns.shape[1]))
        for i in range(patterns.shape[1]):
            for j in range(patterns.shape[1]):
                for u in range(patterns.shape[0]):
                    self.weights[i,j] += (patterns[u,i]-activity)*(patterns[u,j]-activity)
        #for i in range(self.weights.shape[0]):
        #    self.weights[i,i] = 0

    def set_pattern(self, pattern):
        self.pattern = pattern.copy()

    def update_rule(self):
        self.pattern = self.transfer(self.pattern)
        #output = self.transfer(pattern)
        return #output

    def update_rule_async(self):
        for i in range(self.pattern.shape[0]):
            j = np.random.randint(0, self.pattern.shape[0])
            self.pattern[j] = self.sign(np.dot(self.weights[j,:], self.pattern))
            
    def update_rule_sparse(self, theta=1):
        result = self.pattern.copy()
        for i in range(len(self.pattern)):
            result[i] = 0.5 + 0.5 * np.sign(np.dot(self.weights[i,:], self.pattern) - theta)
        self.pattern = result

    def is_fixed_point(self, x):
        self.pattern = x
        self.update_rule()
        if np.all(x == self.pattern):
            return True
        else:
            return False

    def energy(self):
        output = 0
        for i in range(self.pattern.shape[0]):
            for j in range(self.pattern.shape[0]):
                output += self.weights[i,j] * self.pattern[i] * self.pattern[j]
        return -output
```

## 3.2 - Sequential Update

```python
def plot_image(pattern, title):
    plt.imshow(pattern.reshape(32,32))
    plt.title(title)
    plt.show()
```

```python
model = HopfieldNetwork()

patterns = np.loadtxt("pict.dat", delimiter=",", dtype="int")
patterns = patterns.reshape(11, 1024)

targets = patterns[0:3,:]
model.learn_weights(targets)
```

### Check that the three patterns are stable.

```python
data = patterns[0:3,:]
targets = patterns[0:3,:]

model.learn_weights(targets)

for x in data:
    print(model.is_fixed_point(x))
```

### Can the network complete a degraded pattern? Try the pattern _p10_, which is a degraged version of _p1_, or _p11_ which is a mixture of _p2_ and _p3_.

```python
model.set_pattern(patterns[9])
target = patterns[0]
plot_image(model.pattern, "p10")
for i in range(1000):
    before = model.pattern.copy()
    model.update_rule()
    if (np.all(model.pattern == before)):
        break
plot_image(model.pattern, "p10 after batch update rule (" + str(i+1) + " iterations)")
print(np.all(model.pattern == target))

model.set_pattern(patterns[10])
plot_image(model.pattern, "p11")
for i in range(1000):
    before = model.pattern.copy()
    model.update_rule()
    if (np.all(model.pattern == before)):
        break
plot_image(model.pattern, "p11 after batch update rule (" + str(i+1) + " iterations)")
target = patterns[1]
print(np.all(model.pattern == target))
target = patterns[2]
print(np.all(model.pattern == target))
```

### Follow how the energy changes from iteration to iteration when you use the sequential update rule to approach an attractor.

```python
model.set_pattern(patterns[9])
target = patterns[0]
energies = []
for i in range(10000):
    energies += [model.energy()]
    before = model.pattern.copy()
    model.update_rule_async()
    if (np.all(model.pattern == before)):
        energies += [model.energy()]
        break
        
print(np.all(model.pattern == target))
plt.plot(range(len(energies)), energies, label="Energy curve")
model.set_pattern(target)
plt.plot(range(len(energies)), [model.energy() for i in range(len(energies))], label="Target (p1) energy")
plt.xlabel("Iteration step")
plt.ylabel("Energy")
plt.title("Energy curve when training sequentially p10")
plt.legend()
plt.show()
```

## 3.6 - Sparse Patterns

### Try generating sparse patterns with just 10% activity and see how many can be stored for different values of theta.

```python
model = HopfieldNetwork()

n = 200
activity = 0.1
thetas = np.arange(0, 5.5, 1)
nb_patterns = np.arange(1, 50, 2)
y = []

# Iterate over thetas
for theta in thetas:
    stored_patterns = []
    
    # Iterate over number of patterns
    for nb_pattern in nb_patterns:
        value = 0
        
        # Do 10 instances of the same parameters
        for _ in range(10):
            # Generate patterns with activity
            patterns = np.zeros((nb_pattern, n))
            for i in range(nb_pattern):
                idxs = np.random.choice(n, int(activity*n), replace=False)
                for idx in idxs:
                    patterns[i,idx] = 1
            
            # Learn weights with sparse rule
            model.learn_weights_sparse(patterns, activity)
            
            # Check which ones have been remembered
            for i in range(nb_pattern):
                model.set_pattern(patterns[i])
                model.update_rule_sparse(theta)
                if(np.all(model.pattern==patterns[i])):
                    value += 1
            
        value /= 10
        value /= nb_pattern
        stored_patterns += [value]
    
    y += [stored_patterns]

for i, theta in enumerate(thetas):    
    plt.plot(nb_patterns, y[i], label="bias=" + str(theta))

plt.xlabel("Number of patterns")
plt.ylabel("Percentage of remembered patterns")
plt.title("Percentage of remembered patterns by the network in function of the number of patterns and bias w/ activity=" + str(activity))
plt.legend()
plt.show()

for i, theta in enumerate(thetas):    
    plt.plot(nb_patterns[:9], y[i][:9], label="bias=" + str(theta))

plt.xlabel("Number of patterns")
plt.ylabel("Percentage of remembered patterns")
plt.title("Percentage of remembered patterns by the network in function of the number of patterns and bias w/ activity=" + str(activity))
plt.legend()
plt.show()
```
